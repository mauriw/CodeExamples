This code was for a hw assignment for a class, and I'm attaching it because I think its super cool. Basically, using a ton of raw genomic data and some simple data processing, you can figure out pretty clearly what populations people are from and which genomic traits are gender related. It's simple, practical, and beautiful code! 

Since it was for a class weekly assignment its not quite as well documented as some other stuff, but I think the graphs and code are pretty self-explanatory

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from tqdm import tqdm, trange
from sklearn.preprocessing import StandardScaler

%matplotlib inline
sns.set()
```

```python
original = pd.read_csv('p4dataset2019.txt', sep=" ", header=None)
data = original[:] #a copy that will be modified
```

```python
original.head(10)
```

```python
def binarize(data):
    mode = data.mode().iloc[0]
    for i in tqdm(range(3, len(mode))): #go through each column replacing mode values with 0 and others with 1
        data[i] = np.where(data[i] == mode[i], 0, 1)
    return data

data = binarize(data)
```

```python
data.to_csv('binarized.csv') #for future, since binarizing takes a while
```

```python
data.head(10)
```

```python
scaler = StandardScaler(with_std=False)
scaled = pd.DataFrame(scaler.fit_transform(data.iloc[:, 3:].T).T)
scaled.head(5)
```

```python
pop_codes = {
  'ACB': 'African Caribbeans in Barbados',
  'GWD': 'Gambian in Western Divisions in the Gambia',
  'ESN': 'Esan in Nigeria',
  'MSL': 'Mende in Sierra Leone',
  'YRI': 'Yoruba in Ibadan, Nigeria',
  'LWK': 'Luhya in Webuye, Kenya',
  'ASW': 'Americans of African Ancestry in SW USA'
}

def part_b():
    pca = PCA(n_components=2)
    principal = pca.fit_transform(scaled)
    info = {
        'Populations': data[2].replace(pop_codes),
        'First Component' : principal[:, 0],
        'Second Component' : principal[:, 1]
    }
    df = pd.DataFrame(info)
    return sns.relplot(x='First Component', y='Second Component', hue='Populations', data=df)

part_b()
```

```python
def part_d():
    pca = PCA(n_components=3)
    principal = pca.fit_transform(scaled)
    sex = data[1].replace([1, 2], ['Male', 'Female'])
    info = {
        'First component' : principal[:, 0],
        'Third component' : principal[:, 2],
        'Sex' : sex
    }
    df = pd.DataFrame(info)
    return sns.relplot(x='First component', y='Third component', hue='Sex', data=df)

  
part_d()
```

```python
def part_f():
    scaler = StandardScaler(with_std=False)
    scaled = pd.DataFrame(scaler.fit_transform(data.iloc[:, 3:]))
    pca = PCA(n_components=3)
    principal = pca.fit_transform(scaled.T)
    info = {
        'abs(third principal component)' : np.abs(principal[:, 2]),
        'nucleobase index' : scaled.columns
    }
    df = pd.DataFrame(info)
    return sns.relplot(y='abs(third principal component)', x='nucleobase index', edgecolors=None, linewidth=0, data=df)
  
part_f()
```
